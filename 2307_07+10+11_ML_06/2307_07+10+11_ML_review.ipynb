```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

data = load_breast_cancer()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
import numpy as np

tree3 = DecisionTreeClassifier(random_state=0)
params3 = {
    "max_depth":range(1, 6), # 5
    "max_leaf_nodes":range(3, 31), # 28 
    "max_features":np.arange(0.1, 1.1, 0.1) # 10 # í•™ìŠµì‹œ ì‚¬ìš©í•  ì»¬ëŸ¼ì˜ ë¹„ìœ¨
}
print("ì´ ì¡°í•©ìˆ˜:", (5 * 28 *10))

rs = RandomizedSearchCV(tree3, # ëª¨ë¸
                        params3, # í•˜ì´í¼íŒŒë¼ë¯¸í„° í›„ë³´
                        n_iter=60, # í…ŒìŠ¤íŠ¸í•´ë³¼ ì¡°í•©ì˜ ê°œìˆ˜.
                        scoring="accuracy", 
                        cv=4, 
                        n_jobs=-1
                       )
rs.fit(X_train, y_train)
```

```python
print("best score:", rs.best_score_)
print("best parameter:", rs.best_params_)
```
